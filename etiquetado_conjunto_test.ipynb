### Fase de etiquetado para el conjunto de test por separado

Esta fase consistirá en etiquetar las ventanas temporales del IBEX como ejemplos de doble suelo,doble techo y sin patrón para la realización de un conjunto de test por separado.

#### Importamos librerias

```python
import numpy as np
import pandas as pd

import pickle

import matplotlib.pyplot as plt
```

#### Cargamos la serie del IBEX 

```python
with open('benchmark_data.pkl', 'rb') as handle:
    bm = pickle.load(handle)
```

```python
ibex = bm['ibex']
```

```python
ibex.head()
```

#### Instalación e importación de librería mpl_finance
Para realizar los gráficos de velas, utilizaremos la librería mpl_finance.

```python
from mpl_finance import candlestick2_ohlc
```

___

#### Bucle de Etiquetado 

El siguiente código es un bucle que:
1. Elige una posición aleatoria de la serie del IBEX para crear una ventana de 30 días
2. Dibuja el gráfico de velas.
3. Pide al usuario que introduzca un número para etiquetar la gráfica.

Primero hay que elegir el tamaño de la ventana y el número de ejemplos que se quieren generar.  
Se tiene la opción de descartar algún ejemplo si no se quiere incluir (ej. posiblemente ambiguo)

```python
window =30
n = 200 
```

```python
etiquetas = np.zeros(n)
idfechas = np.zeros(n)
for i in range(n):
    idx = np.random.randint(0, ibex.shape[0] - window)
    idfechas[i] = idx
    windata = ibex.iloc[idx:idx+window]
    fig, ax = plt.subplots(figsize=(6,4))
    _ = candlestick2_ohlc(ax, windata.open, windata.high,
                         windata.low, windata.close,
                         colorup='g', colordown='r', width=0.66, alpha=0.8)
    plt.show()
    print(f'Ejemplo {i+1} de {n}')
    print("Etiquete la siguiete grafica como:")
    print(" [1] Doble suelo, [2] Doble techo , [0] Sin patrón, [3] Descartar ejemplo")
    opcion_usuario = input()
    if opcion_usuario >= '0' and opcion_usuario <= '3':
        etiquetas[i] = opcion_usuario   
    else:
        print('Error en la opción introducida. Ejemplo descartado')
        etiquetas[i] = 3
```

```python
doble_suelo_techo_sin_patron_2_labels = pd.Series(etiquetas, index=ibex.index[idfechas.astype(np.int)])
```

Descartamos las etiquetas descartadas

```python
doble_suelo_techo_sin_patron_2_labels = doble_suelo_techo_sin_patron_2_labels[doble_suelo_techo_sin_patron_2_labels!=3]
```
